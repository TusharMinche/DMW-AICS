```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/digra/Downloads/Tusahr/Titanic.csv")
df
```

```python
df.head()
```

```python
# 1. Handling Missing Values
# Fill missing values in Age column with mean value
# Using dictionary-like syntax to set values inplace
df['age'] = df['age'].fillna(df['age'].mean())

# Alternatively, using the fillna method directly on the DataFrame without chaining
df.fillna({'age': df['age'].mean()}, inplace=True)
```

```python
# Drop rows with missing values in Embarked column
df.dropna(subset=['embarked'], inplace=True)
df.head()
```

```python
# 2. Categorical Encoding
# Convert Sex and Embarked columns to numerical values
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked'])
df.head()
```

```python
# 3. Feature Scaling
# Normalize Fare column
df['fare'] = (df['fare'] - df['fare'].min()) / (df['fare'].max() - df['fare'].min())
df.head()
```

```python
# 4. Feature Generation
# Create a new feature FamilySize by combining SibSp and Parch
df['FamilySize'] = df['sibsp'] + df['parch'] + 1
df.head()
```

```python
# 5. Feature Selection
# Drop unnecessary columns
df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df.head(3)
```
