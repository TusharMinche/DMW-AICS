```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/ADMIN/Desktop/CLG/AICS/Titanic.csv")
df
```

```python
df.head()
```

```python
# 1. Handling Missing Values
# Fill missing values in Age column with mean value
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.head()
```

```python
# Drop rows with missing values in Embarked column
df.dropna(subset=['Embarked'], inplace=True)
df.head()
```

```python
# 2. Categorical Encoding
# Convert Sex and Embarked columns to numerical values
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
df.head()
```

```python
# 3. Feature Scaling
# Normalize Fare column
df['Fare'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())
df.head()
```

```python
# 4. Feature Generation
# Create a new feature FamilySize by combining SibSp and Parch
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
# 5. Feature Selection
# Drop unnecessary columns
df.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df.head(3)
```
